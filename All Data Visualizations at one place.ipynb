```python
# import 
import matplotlib.pyplot as plt
import seaborn as sns

# set styles for sns and plt
plt.style.use('seaborn-whitegrid')
sns.set_style('white')

# for supress warnings
import warnings 
warnings.filterwarnings('ignore')

# for HD Visualizations
%config InlineBackend.figure_format='retina'
```

<h2 id ="1.-Scatter-plot">1. Scatter plot</h2>

> Scatteplot is a classic and fundamental plot used to study the relationship between two numerical variables.

```python
df = sns.load_dataset('iris')
df.head()
```

```python
plt.figure(figsize=(7,5), dpi = 100, facecolor='w', edgecolor='k')
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = df,hue='species',style='species')
plt.show();
```

<h2 id = "2. Scatter plot with linear regression line of best fit">2. Scatter plot with linear regression line of best fit</h2>

>If you want to understand how two variables change with respect to each other, the line of best fit is the way to go. The below plot shows how the line of best fit differs amongst various groups in the data.

```python
grid_obj = sns.lmplot(x = 'sepal_length', y ='petal_length', data = df,hue = 'species',height=6,aspect=2.0,robust=True, palette='tab10',
                     scatter_kws = dict(s=150,linewidths=.5, edgecolors='black'),markers='*')

grid_obj.set(xlim = (4.0, 8.0), ylim=(0,15))
plt.title("Scatterplot with line of best fit grouped by species", fontsize=20)
plt.show();
```

<h2 id ="3.-Strip-plot">3. Strip-plot</h2>

> A Strip-plot in seaborn is typically a Categorical-scatter plot used to visualize the distribution of a continuous variable across different categories where the data points are `"jittered"` horizontally to prevent overplotting within each category.

```python
plt.figure(figsize=(7,5), dpi = 100, facecolor='w', edgecolor='k')
sns.stripplot(x = 'sepal_width', y = 'species',data = df, hue='species', )
plt.show();
```

<h2 id ="4.-Box-plot">4. Box-plot</h2>

> Box-plot also know for box-and-whisker plot is used to visualize the distribution of the quantitative data.

**Components of a Box-plot**
* **Box**: Represents the interquartile range (IQR), which encompasses the middle 50% of the data.
* **Whisker**: Extending from the box represents the bottom and the top 25% of the data. They usually reach upto 1.5 times the IQR.
* **Median Line**: Runs Horizontal Through the box, representing 50th percent of the data.
* **Outliers**: Data points falling outside the whiskers.

```python
plt.figure(figsize=(7,5), dpi = 100, facecolor='w', edgecolor='k')
sns.boxplot(x = 'sepal_width', data = df,hue='species', orient='v', color='aqua')
plt.show();
```

<h2 id ="5.-Count-plot">5. Count-plot</h2>

> A count plot is a powerful visualization tool to view the frequency of observations in each category of a categorical variable. This is typically a bar-plot but specifically designed for categorical data.

```python
df_1 = sns.load_dataset('tips')
df_1.head()
```

```python
plt.figure(figsize=(7,5), dpi = 100, facecolor='w', edgecolor='k')
sns.countplot( x = 'day', data = df_1, hue='time', orient='v',palette='tab10',width=0.6, saturation=1.5)
plt.show();
```

<h2 id ="6.-Correlation Heat Map">6. Correlation Heat Map </h2>

> Correlation Heat Map is used to visually see the correlation metric between all possible pairs of numeric variables in a given dataframe (or 2D array).

```python
df_2 = sns.load_dataset('diamonds')
df_2
```

```python
plt.figure(figsize=(10,7),dpi =100, facecolor='w',edgecolor='k')

sns.heatmap(df_2.corr(),xticklabels=df_2.corr().columns,yticklabels=df_2.corr().columns
            ,cmap='RdYlGn',center=0,annot =True)

plt.title('Correlation Heat Map')
plt.xticks(fontsize = 15)
plt.yticks(fontsize=15)
plt.show();
```

<h2 id ="7.-Pair Plot">7. Pair Plot </h2>

> Pairwise plot is a favorite in exploratory analysis to understand the relationship between all possible pairs of numeric variables. It is a must have tool for bivariate analysis.

```python
plt.figure(figsize=(10,7),dpi =100, facecolor='w',edgecolor='k')

sns.pairplot(df_2, hue="cut", plot_kws=dict(s=100, edgecolor="white"))

plt.show();
```
